```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.optimize import curve_fit 
```

```python
def exp_law(x, a, b):
    return a*np.exp(b*x)
```

```python
def power_law(x, a, b):
    return a*x**b
```

# Curve fits for BC

```python
x = np.array([i for i in range(0, len(BC))])
number_of_days_into_the_future = 10 
x_peredictions = np.array([i for i in range(0, len(BC)+number_of_days_into_the_future)])
BC_exp, BC_exp_cov = curve_fit(exp_law, x/len(BC), np.array(BC))
BC_exp_ans = (BC_exp[0]*np.exp(BC_exp[1]*x_peredictions/len(BC))) 
BC_pow, BC_pow_cov = curve_fit(power_law, x/len(BC), np.array(BC))
BC_pow_ans = BC_pow[0]*(x_peredictions/len(BC))**BC_pow[1]
print(BC_exp)
print(BC_exp_cov)
print(BC_pow)
print(BC_pow_cov)
```

```python
plt.figure(figsize=(10,10))
plt.plot(BC_pow_ans, '-*g', label ="power law fit")
plt.plot(BC_exp_ans, '-b', label ="exponential fit")
plt.plot(BC, 'or', label ="BC")
plt.legend() 
plt.grid()
#plt.plot(Ontario, '+b')
```

```python
x = np.array([i for i in range(0, len(France))])
number_of_days_into_the_future = 10 
x_peredictions = np.array([i for i in range(0, len(France)+number_of_days_into_the_future)])
France_exp, France_exp_cov = curve_fit(exp_law, x/len(France), np.array(France))
France_exp_ans = (France_exp[0]*np.exp(France_exp[1]*x_peredictions/len(France))) 
France_pow, France_pow_cov = curve_fit(power_law, x/len(France), np.array(France))
France_pow_ans = France_pow[0]*(x_peredictions/len(France))**France_pow[1]
```
